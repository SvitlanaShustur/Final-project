## Побудова регресійної моделі для прогнозування суми покупки

### 1. Мета моделювання

Метою моделювання було оцінити можливість прогнозування **Gross Amount** (суми покупки до застосування знижки) на основі характеристик клієнта та транзакції, а також перевірити вплив вікової групи.

В задачі було зазначено, що маэ бути застосована модель лінійної рекресії.

### 2. Підготовка даних

#### 2.1 Формування ознак

До моделі було включено такі змінні:

- **Gender** (стать)
- **Age Group** (вікова група)
- **Product Category** (категорія товару)
- **Discount Availed** (факт застосування знижки)
- **Purchase Method** (спосіб оплати)
- **Location** (місто покупки)
- **Purchase Date** (дата покупки)

Цільова змінна — **Gross Amount**.

#### 2.2 Обробка дати

Колонку `Purchase Date` перетворено у числові ознаки:

- `purchase_year` — рік покупки  
- `purchase_month` — місяць  
- `purchase_dayofweek` — день тижня  

Оригінальну дату видалено.

#### 2.3 Обробка вікової групи

`Age Group` є порядковою змінною (ordinal variable), тому було:

1. Задано логічний порядок вікових категорій.
2. Створено числову ознаку `Age_Numeric` через порядкове кодування (`cat.codes`).

Це дозволило моделі інтерпретувати перехід до старшої вікової категорії.

#### 2.4 Категоріальні змінні

Для текстових змінних застосовано **OneHotEncoder (one-hot кодування)**, що перетворює категорії на набір 0/1 ознак.

#### 2.5 Очистка даних

Було видалено рядки з пропущеними значеннями (`dropna`) для забезпечення коректного навчання моделі (на всяк випадок, т.я. дані чисті).
### 3. Побудова моделі

Модель реалізована через:

- **ColumnTransformer** — для окремої обробки категоріальних та числових змінних;
- **Pipeline** — для автоматизації етапів підготовки та навчання;
- **LinearRegression** — як базову регресійну модель.

Дані розділено на:
- **train** (80%)
- **test** (20%)

### 4. Виявлення витоку даних (Data Leakage)

Початково до моделі було включено `Net Amount`, що призвело до надзвичайно високого:

- **R² ≈ 0.997**
- **MAE ≈ 57**

Аналіз показав наявність **data leakage (витоку даних)**, оскільки Gross Amount математично пов’язаний з Net Amount

Після виключення `Net Amount` модель була перенавчена.

### 5. Фінальні результати (без витоку даних)

- **Коефіцієнт Age_Numeric:** 7.95  
- **R²:** 0.0417  
- **MAE:** 1434.88  

### 6. Інтерпретація результатів

#### Вплив вікової групи

Коефіцієнт 7.95 означає:

> Перехід до наступної вікової категорії збільшує прогнозовану суму покупки в середньому на 7.95 грошових одиниць за інших рівних умов.

Ефект є позитивним, але незначним.

#### Якість моделі

- **R² = 0.0417** означає, що модель пояснює лише 4% варіації суми покупки.
- **MAE ≈ 1435** демонструє обмежену точність прогнозу.

Це свідчить про слабкий зв’язок між використаними характеристиками та сумою покупки.

### 7. Висновки

1. Вікова група має слабкий позитивний вплив на суму покупки.
2. Демографічні характеристики та транзакцій не дозволяють точно прогнозувати доходи.
3. Основна варіація ймовірно визначається факторами, відсутніми в датасеті (конкретні товари, історія покупок, маркетингові активності).
4. Усунення витоку даних дозволило отримати реалістичну оцінку моделі.
